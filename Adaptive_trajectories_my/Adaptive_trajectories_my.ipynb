# Обработка plagiarism_comparisonSet в solutions

```python
# -1 - данных нет  (потом добавлю уже в объединенной табличке, тк там есть год, а тут мне лень выделять его)
# 0 - плагиата нет
# 1 - плагиат больше 0, меньше 50
# 2 - плагиат больше 50, меньше 80
# 3 - плагиат больше 80

Plagiat = []
for numSol in range(len(solutions)):
  if (solutions.plagiarism_comparisonSet[numSol] == '{}')or(pd.isnull(solutions.plagiarism_comparisonSet[numSol])):
    Plagiat.append(0)
  else:
    A=[]
    k = 0
    text = solutions.plagiarism_comparisonSet[numSol]
    for i in range(len(text)):
      if text[i] == 'p':
        k = 1
      if text[i] == 'r' and k==1:
        k += 1
      if text[i] == 'o' and k==2:
        k += 1
      if text[i] == 'b' and k==3:
        k += 1
      if text[i] == '"' and k==4:
        k += 1
      if text[i] == ":" and k==5:
        k += 1
      if text[i] == " " and k==6:
        k = 0
        count = 1
        A.append('')
        while '0'<=text[i+count]<='9' or text[i+count] == '.':
          A[-1]+=text[i+count]
          count+=1
        i+=count

    for i in range(len(A)):
      if A[i]=='':
        A[i] = 0
      A[i] = float(A[i])

    inTab = 0
    for i in A:
      if i == 0 and inTab==0:
        inTab = 0
      elif i>0 and i<0.50 and inTab<1:
        inTab = 1
      elif i>=0.50 and i<0.80 and inTab<2:
        inTab = 2
      elif i>=0.80 and inTab<3:
        inTab = 3
    Plagiat.append(inTab)
```

```python
solutions['isPlag'] = Plagiat
```

```python
time = [] #30+31+30+31+31+28+31+30
for i in range(len(data_passed)):
  if data_passed.year[i] == '2022':
    if data_passed.month[i] == '09':
      time.append(int(data_passed.day[i]))
    elif data_passed.month[i] == '10':
      time.append(30+int(data_passed.day[i]))
    elif data_passed.month[i] == '11':
      time.append(30+31+int(data_passed.day[i]))
    elif data_passed.month[i] == '12':
      time.append(30+31+30+int(data_passed.day[i]))

  elif data_passed.year[i] == '2023':
    if data_passed.month[i] == '01':
      time.append(30+31+30+31+int(data_passed.day[i]))
    elif data_passed.month[i] == '02':
      time.append(30+31+30+31+31+int(data_passed.day[i]))
    elif data_passed.month[i] == '03':
      time.append(30+31+30+31+31+28+int(data_passed.day[i]))
    elif data_passed.month[i] == '04':
      time.append(30+31+30+31+31+28+31+int(data_passed.day[i]))

data_passed['time'] = time
```
